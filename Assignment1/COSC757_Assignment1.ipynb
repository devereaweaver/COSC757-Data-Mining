# Tests of Normality

Graphically, none of the variables in the dataset appear to be normally distributed.

However, we shouldn't only rely on graphical tools, we should also use support findings with statistical tests as well.
Doing so will improve our analysis.

To help support our visual evidence, we'll also conduct a few statistical tests for normality.
These include:
1. Shapiro-Wilk test
2. Kolmogorov-Smirnov test

These two tests are used often in statistical inference to determine whether variables empirically come from a normal distribution.

They all have the following hypotheses:
$$H_0: (X_1, X_,2 \dots, X_n) \sim Norm(\mu, \sigma)$$
$$H_A: (X_1, X_,2 \dots, X_n) \nsim Norm(\mu, \sigma)$$

where $X_1, X_2, \dots, X_n$ are samples of a given variable.

Thus, computationally, a "small" p-value means that we'll assume the data is not normally distributed and a "large" p-value means that we'll assume the data is normally distributed.

All three are computed differently; however, the results for each test will *generally* agree, although not always.

 **TODO: Move the "examples" of using the tests, before actually doing the tests on the real data."**

```python
# conduct shapiro-wilk test of normality on all vars
from scipy.stats import shapiro, kstest

print("Tests of Normality:")
for name in data.columns:
  print(f"{name}: \n {shapiro(data[name])} \n {kstest(data[name], 'norm')} \n")
```

Based on the results of the tests, we'll assume that none of the variables come from a normal distribution.

## Attempt to normalize some variables

### 1. min-max

We'll attempt to normalize variables using min-max normalization.


```python
# function that will min-max normalize the input data
def min_max(x):
  range = max(x) - min(x)
  return (x - min(x))/(range)
```

```python
# create min-max normalized to dataframe
min_max_df = pd.DataFrame()

for name in data.columns:
    min_max_df['Cement'] = min_max(data['Cement'])
    min_max_df['Blast_Furnance_Slag'] = min_max(data['Blast_Furnance_Slag'])
    min_max_df['Fly_Ash'] = min_max(data['Fly_Ash'])
    min_max_df['Water'] = min_max(data['Water'])
    min_max_df['Superplasticizer'] = min_max(data['Superplasticizer'])
    min_max_df['Coarse_Aggregate'] = min_max(data['Coarse_Aggregate'])
    min_max_df['Fine_Aggregate'] = min_max(data['Fine_Aggregate'])
    min_max_df['Age'] = min_max(data['Age'])
    min_max_df['Compressive_Strength'] = min_max(data['Compressive_Strength'])
```

### 2. Z-score

Do the same thing but using a z-score function instead.

```python
# z-score function
def z_score(x):
  mean = np.mean(x)
  sd = np.std(x)
  return (x - mean)/sd
```

```python
# create a new z-score dataframe
z_score_df = pd.DataFrame()

for name in data.columns:
    z_score_df['Cement'] = z_score(data['Cement'])
    z_score_df['Blast_Furnance_Slag'] = z_score(data['Blast_Furnance_Slag'])
    z_score_df['Fly_Ash'] = z_score(data['Fly_Ash'])
    z_score_df['Water'] = z_score(data['Water'])
    z_score_df['Superplasticizer'] = z_score(data['Superplasticizer'])
    z_score_df['Coarse_Aggregate'] = z_score(data['Coarse_Aggregate'])
    z_score_df['Fine_Aggregate'] = z_score(data['Fine_Aggregate'])
    z_score_df['Age'] = z_score(data['Age'])
    z_score_df['Compressive_Strength'] = z_score(data['Compressive_Strength'])
```

### 3. Decimal Scaling



```python
x = np.abs(data['Cement'])
x = np.max(x)
x_str = str(x)
len(x_str) - 1
```

```python
def decimal_scale(x):
  a = np.abs(x)
  a = np.max(x)
  a_str = str(a)
  d = len(a_str) - 1
  return (x)/(10**d)
```

```python
# create decimal scaled dataframe
decimal_scaled_df = pd.DataFrame()

for name in data.columns:
    decimal_scaled_df['Cement'] = decimal_scale(data['Cement'])
    decimal_scaled_df['Blast_Furnance_Slag'] = decimal_scale(data['Blast_Furnance_Slag'])
    decimal_scaled_df['Fly_Ash'] = decimal_scale(data['Fly_Ash'])
    decimal_scaled_df['Water'] = decimal_scale(data['Water'])
    decimal_scaled_df['Superplasticizer'] = decimal_scale(data['Superplasticizer'])
    decimal_scaled_df['Coarse_Aggregate'] = decimal_scale(data['Coarse_Aggregate'])
    decimal_scaled_df['Fine_Aggregate'] = decimal_scale(data['Fine_Aggregate'])
    decimal_scaled_df['Age'] = decimal_scale(data['Age'])
    decimal_scaled_df['Compressive_Strength'] = decimal_scale(data['Compressive_Strength'])
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
# select features
features = data.iloc[:, :-1]
features
```

```python
# select target
target = data.iloc[:, -1]
target
```

```python
# stepwise regresssion
sfs = SequentialFeatureSelector(linear_model.LinearRegression(),
                                k_features=8,
                                forward = True,
                                scoring = "neg_mean_squared_error",
                                cv = None)

# fit select features
selected_features = sfs.fit(features, target)
```

```python
# check which features were used
selected_features.k_feature_names_
```

```python
# do the test/train split thing
x_train, x_test, y_train, y_test = train_test_split(
    data.iloc[:, :-1],
    target,
    test_size = .2,
)
```
